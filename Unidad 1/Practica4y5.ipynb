# Practica 4

tablas de frecuencia

```python
#creamos una lista vacia  que representa la edad de las 100 personas
import random
edades=[]

#Hacemos un ciclo para general 100 numero aleatorios que representan la edad de las personas
#generamos un numero aleatorio con la funcion randint y lo guardamos en la variable
#numero alterno
for i in range(100):
  numero_aleatorio=random.randint(18,70)
#agregamos el numero aleatorio a la lista de valores
  edades.append(numero_aleatorio)
#imprimimos la lista de edades
print(edades)
```

```python
import pandas as pd

#crear un dataframe de pandas con los numeros aleatorios
df=pd.DataFrame(edades,columns=['edades'])
df.head()
```

```python
#calcular la tabla de distribucion de frecuencias
tabla_frecuencias=df['edades'].value_counts().reset_index()
tabla_frecuencias.columns =['edades', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='edades')

#imprimir la tabla
print("tabla de frecuencias")
tabla_frecuencias
```

```python
tabla_frecuencias['Frecuencia relativa'] = tabla_frecuencias['Frecuencia'] /100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec porcentual'] = tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

# Regla de sturges

```python
#Es una regla que sirve para calcular el numero de clases o intervalos idoneos que se debe dividir un
#conjunto de datos

#C=1 + logbase2(N)

#importamos la libreria que permite funciones matematicas

import math

#usar la lista edades
print(edades)
```

```python
#crear un dataframe con las edades
df=pd.DataFrame({'Edad' : edades})
df
```

```python
#calcular el numero de intervalos utilizando la regla de sturges

n=len(edades)
k= int(1 + math.log2(100))
print("Valor de n= ", n , " \nNumero de intervalos =",k)
```

```python
from ast import increment_lineno
#crear la tabla de frecuencia

tabla_frecuencias = pd.cut( df['Edad'],bins=k, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns =['Clase','Frec']
tabla_frecuencias
```

```python
tabla_frecuencias = tabla_frecuencias.sort_values(by='Clase')
tabla_frecuencias
```

```python
#generar columnas siguientes
tabla_frecuencias['Frec Rel'] = tabla_frecuencias['Frec'] /n
tabla_frecuencias['Frec Porc'] = tabla_frecuencias['Frec']/n*100
tabla_frecuencias['Frec Ac'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac'] = tabla_frecuencias['Frec Rel'].cumsum()
tabla_frecuencias['Frec Porc Ac'] = tabla_frecuencias['Frec Porc'].cumsum()

#mostrar la tabla de frecuencia
print(tabla_frecuencias)
```
