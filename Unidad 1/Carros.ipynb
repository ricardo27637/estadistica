```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
#iniciar el archivo
autos=pd.read_csv("autos2.csv", encoding="ISO-8859-1")
```

```python
#mostrar los registros del codigo

autos.head()
```

```python
#mostrar la informacion del archivo

autos.info()
```

```python
autos.price.min()
```

```python
autos.price.describe()
```

```python
#dibujar el histograma
#calcular la media y guardarla en una variable

media=autos.price.mean()

plt.pyplot.hist(autos.price, bins=11)
plt.pyplot.title("Histograma de precios")
plt.pyplot.axvline(media, color='y',linestyle='dashed',linewidth=2, label=f'Media={media:.2f}')
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("frecuencia")
```

```python
#histograma de la columna estilos

plt.pyplot.hist(autos['body-style'], bins=3)
plt.pyplot.title("Histograma de estilos")
plt.pyplot.xlabel("Estilo")
plt.pyplot.ylabel("frecuencia")
```

```python
#diagrama de cajas

sns.boxplot(x="body-style", y="price",data=autos)
```

```python
#diagrama de cajas drive wells

sns.boxplot(x="drive-wheels", y="price",data=autos)
```

```python
#diagrama de cajas engine-location

sns.boxplot(x="engine-location", y="price",data=autos)
```
