#distribución de Poisson.

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Datos del problema
media = 5

# a) ¿Cuál es la probabilidad de que el número de personas ingresadas a una unidad de cuidados intensivos en un hospital particular, en un día particular, sea dos?

probabilidad_dos = poisson.pmf(2, media)
print(f"La probabilidad de que ingresen exactamente dos personas es: {probabilidad_dos:.4f}")

# b) ¿Es probable que x exceda de 10?

# Calculamos la probabilidad de que x sea mayor que 10
probabilidad_mayor_diez = 1 - poisson.cdf(10, media)
print(f"La probabilidad de que ingresen más de diez personas es: {probabilidad_mayor_diez:.4f}")

# si es probable:
# La probabilidad de que x exceda de 10 es {probabilidad_mayor_diez:.4f}, que es un valor
# relativamente bajo. Esto sugiere que no es muy probable que más de 10 personas
# ingresen a la UCI en un día, dado que la tasa promedio es de 5 personas por día.

# Visualización de la distribución de Poisson
x = np.arange(0, 15)
pmf = poisson.pmf(x, media)

plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='skyblue')
plt.xlabel('Número de personas ingresadas (x)')
plt.ylabel('Probabilidad P(X=x)')
plt.title(f'Distribución de Poisson con media = {media}')
plt.xticks(x)
plt.grid(axis='y', linestyle='--')
plt.show()
```

#distribución binomial

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# Datos del problema
n = 1000  # Número de motores (ensayos)
p = 0.001 # Probabilidad de que un motor sea defectuoso (probabilidad de éxito)

# a) ¿Cuál es la probabilidad de que ninguno sea defectuoso?
k = 0  # Número de éxitos (motores defectuosos)
probabilidad_ninguno_defectuoso = binom.pmf(k, n, p)
print(f"La probabilidad de que ninguno sea defectuoso es: {probabilidad_ninguno_defectuoso:.4f}")

# b) ¿Cuál es la probabilidad de que tres sean defectuosos?
k = 3
probabilidad_tres_defectuosos = binom.pmf(k, n, p)
print(f"La probabilidad de que tres sean defectuosos es: {probabilidad_tres_defectuosos:.4f}")

# c) ¿Cuál es la probabilidad de que cuatro sean defectuosos?
k = 4
probabilidad_cuatro_defectuosos = binom.pmf(k, n, p)
print(f"La probabilidad de que cuatro sean defectuosos es: {probabilidad_cuatro_defectuosos:.4f}")

# Visualización de la distribución binomial (para un rango pequeño de defectos)
k_values = np.arange(0, 10)
pmf_values = [binom.pmf(k, n, p) for k in k_values]

plt.figure(figsize=(10, 6))
plt.bar(k_values, pmf_values, color='lightcoral')
plt.xlabel('Número de motores defectuosos (k)')
plt.ylabel('Probabilidad P(X=k)')
plt.title(f'Distribución Binomial (n={n}, p={p})')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--')
plt.show()
```
