```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
import seaborn as sns
from scipy.stats import pearsonr

# Subir los archivos
uploaded = files.upload()

# Cargar los datos
df1 = pd.read_csv("grupo1.csv")
df2 = pd.read_csv("grupo2.csv")
df3 = pd.read_csv("grupo3.csv")

# Función para realizar el análisis completo de un dataframe
def analizar_grupo(df, grupo_num):
    print(f"\n{'='*50}")
    print(f"ANÁLISIS PARA GRUPO {grupo_num}")
    print(f"{'='*50}\n")

    # Exploración inicial
    print("1. EXPLORACIÓN INICIAL DE DATOS")
    print("Primeras filas del dataset:")
    print(df.head())
    print("\nDescripción estadística:")
    print(df.describe())

    # Histograma de calificaciones
    plt.figure(figsize=(8, 4))
    plt.hist(df['Calificacion'], bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Distribución de Calificaciones - Grupo {grupo_num}')
    plt.xlabel('Calificación')
    plt.ylabel('Frecuencia')
    plt.grid(True, alpha=0.3)
    plt.show()

    # Modelo de regresión lineal
    print("\n2. MODELO DE REGRESIÓN LINEAL")
    X = df[["HorasEstudio"]]
    y = df["Calificacion"]

    lm = LinearRegression()
    lm.fit(X, y)

    print(f"Ecuación de regresión: Calificación = {lm.intercept_:.2f} + {lm.coef_[0]:.2f} * HorasEstudio")
    print(f"Pendiente (coeficiente): {lm.coef_[0]:.2f}")
    print(f"Intercepto: {lm.intercept_:.2f}")
    print(f"R-cuadrado (coeficiente de determinación): {lm.score(X, y):.4f}")

    # Coeficiente de correlación de Pearson
    corr, p_value = pearsonr(df["HorasEstudio"], df["Calificacion"])
    print(f"\n3. COEFICIENTE DE CORRELACIÓN DE PEARSON: {corr:.4f}")
    print(f"Valor p: {p_value:.4f}")

    # Visualización
    print("\n4. VISUALIZACIÓN DE RESULTADOS")
    plt.figure(figsize=(10, 6))
    sns.regplot(x='HorasEstudio', y='Calificacion', data=df,
                scatter_kws={'color': 'blue', 'alpha': 0.5},
                line_kws={'color': 'red'})
    plt.title(f'Relación entre Horas de Estudio y Calificación - Grupo {grupo_num}')
    plt.xlabel('Horas de Estudio')
    plt.ylabel('Calificación')
    plt.grid(True)
    plt.show()

    # Interpretación de resultados
    print("\n5. INTERPRETACIÓN DE RESULTADOS")
    print("1. Interpretación del coeficiente de regresión:")
    print(f"   Por cada hora adicional de estudio, se espera un aumento de {lm.coef_[0]:.2f} puntos en la calificación.")

    print("\n2. Interpretación del coeficiente de correlación:")
    if corr > 0.7:
        print("   Existe una fuerte correlación positiva entre las variables.")
    elif corr > 0.3:
        print("   Existe una moderada correlación positiva entre las variables.")
    elif corr > 0:
        print("   Existe una débil correlación positiva entre las variables.")
    elif corr == 0:
        print("   No existe correlación lineal entre las variables.")
    else:
        print("   Existe una correlación negativa entre las variables.")

    print("\n3. Significancia estadística del modelo:")
    if p_value < 0.05:
        print("   El modelo es estadísticamente significativo (p < 0.05).")
    else:
        print("   El modelo NO es estadísticamente significativo (p ≥ 0.05).")

    print("\n4. Confiabilidad del modelo para predicciones:")
    r_squared = lm.score(X, y)
    if r_squared > 0.7:
        print(f"   El modelo es bastante confiable (R² = {r_squared:.2f}), explica más del 70% de la variabilidad.")
    elif r_squared > 0.5:
        print(f"   El modelo es moderadamente confiable (R² = {r_squared:.2f}), explica más del 50% de la variabilidad.")
    else:
        print(f"   El modelo tiene limitada confiabilidad (R² = {r_squared:.2f}), explica menos del 50% de la variabilidad.")

# Realizar análisis para cada grupo
analizar_grupo(df1, 1)
analizar_grupo(df2, 2)
analizar_grupo(df3, 3)

print("\nCONCLUSIONES GENERALES:")
print("1. Los resultados muestran las diferencias en la relación entre horas de estudio y calificaciones en los tres grupos.")
print("2. Se observa que en algunos grupos la relación es más fuerte que en otros.")
print("3. El análisis de correlación y regresión permite cuantificar estas diferencias.")
print("4. Los gráficos ayudan a visualizar la relación y la distribución de los datos.")
print("5. La significancia estadística indica si las relaciones observadas son probablemente reales o podrían deberse al azar.")
```
